```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',100)
```

```python
data = 'HR_Employee_Attrition_Data.csv'
```

```python
df =pd.read_csv(data)
```

```python
print("Columns and unique value counts and Feature Data types")
data_summary = pd.DataFrame(df.dtypes)
data_summary['uCount'] =0
columns = df.columns.to_list()
for i in range(len(columns)):
    data_summary.iloc[i,1] = len(df[columns[i]].unique())
data_summary.rename(columns={0:'dtype'},inplace=True)
data_summary.T
```

```python
df.Attrition.value_counts()['No']
df.Attrition.value_counts()['Yes']
```

```python
print(min(df.DailyRate.unique()),max(df.DailyRate.unique()))
```

```python
age_dist = pd.DataFrame()
age_dist['age'] = df['Age']
age_dist['attrition'] = df['Attrition']
age_dist['DailyRate'] = df['DailyRate']
age_dist
```

```python
#group_one = 18 - 24
#group_two = 25 - 30
#group_thre = 31- 36
#group_four = 37-42
#group_five = 43-48
#group_six = 49-54
#group_seven = 55-60
age_dist['age_bucket']="a"
for i in range(len(age_dist)):
    x = age_dist.iloc[i,0]
    if x>=18 and x<=24:
        age_dist.iloc[i,3] ="18-24"
    elif x>24 and x<=30:
        age_dist.iloc[i,3] ="25-30"
    elif x>30 and x<=36:
        age_dist.iloc[i,3] ="31-36"
    elif x>36 and x<=42:
        age_dist.iloc[i,3] = "37-42"
    elif x>42 and x<=48:
        age_dist.iloc[i,3] ="43-48"
    elif x>48 and x<=54:
        age_dist.iloc[i,3] ="49-54"
    else:
        age_dist.iloc[i,3]= "54+"
```

```python
plot_one = data_summary[data_summary['uCount']<=5]
plot_one.T
```

```python
sns_cols = list(plot_one[plot_one['uCount']>1]['uCount'].keys())
```

```python
sns_cols = np.array(sns_cols).reshape(5,3)
sns_cols
```

```python
fig,axes = plt.subplots(5,3,figsize=(25,30))
for i in range(5):
    for j in range(3):
        if sns_cols[i][j]=="Attrition":
            sns.countplot(df[sns_cols[i][j]],ax=axes[i,j],palette="Blues")
        else:
            #sns.countplot(df[sns_cols[i][j]],hue=df['Attrition'],ax=axes[i,j])
            axes[i][j].set_xlabel(sns_cols[i][j],fontsize=15)
            axes[i][j].set_ylabel("Attrition Count")
            axes[i][j].set_ylim(0,max(df[sns_cols[i][j]].value_counts())+500)
            p = sns.countplot(df[sns_cols[i][j]],hue=df['Attrition'],ax=axes[i,j],palette="Blues")
            for x in p.patches:
                p.annotate(str(format(x.get_height()/len(df)*100,'.2f'))+"%",(x.get_x()+x.get_width() / 2.,x.get_height()),
                           ha='center',va='center',xytext=(0,10),textcoords='offset points',fontsize=12,fontweight='normal')
    
plt.show()                                   
```
